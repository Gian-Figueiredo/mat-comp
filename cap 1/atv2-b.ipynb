## Funções importantes

```python
def plotar_basico(x, y, function, title="", xlabel="Quantidade de carros", ylabel="Renda (em mil reais)"):
    plt.plot(x, function)
    plt.scatter(x, y)
    plt.xlim(0, 7)
    plt.ylim(0,7)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()
```

# Gradiente descedente nos dois parâmtro da regressão linear

Utilizando os mesmo dados

```python
x = np.arange(1, 6)
y = [3, 1, 5, 4, 4]
learning_rate = 0.1

plt.scatter(x, y)
plt.xlim(0, 7)
plt.ylim(0,7)

plt.title("Dados da coleta")
plt.xlabel("Renda (em mil reais)")
plt.ylabel("Quantidade de carros")

plt.show()
```

Agora precisamos descobrir não só o intercept como também o bias. O processo não muda, porém haverá duas funções residual, uma para o bias e outra para o intercept. Consideraremos o bias = 0.5 e intercept = 1, como chute inicial

```python
bias = 1.5
intercept = 1
```

Criamos a função previsto com base nos valores chutados

```python
previsto = lambda intercept, bias: np.array([bias * i + intercept for i in x])
esperado = np.array(y)
```

```python
plotar_basico(x, y, previsto(intercept, bias), title="Reta inicial")
```

Criamos a função residual, que é a soma dos quadrados da diferença entre o previsto e o esperado

```python
residual_bias = lambda bias : np.square(previsto(intercept, bias) - esperado)
residual_intercept = lambda intercept : np.square(previsto(intercept, bias) - esperado)
```

Para encontrar o mínimo, encontraremos a soma dos gradientes de intercept e bias separadamente

```python
sum_gradiente_intercep = np.sum(np.gradient(residual_intercept(intercept), x))
sum_gradiente_bias = np.sum(np.gradient(residual_bias(bias), x))
display(sum_gradiente_bias, sum_gradiente_intercep)
```

Agora, multiplicaremos pelo coeficiente de aprendizado e modificaremos os parametros

```python
step_intercept = sum_gradiente_intercep * learning_rate
step_bias = sum_gradiente_bias * learning_rate
intercept -= step_intercept
bias -= step_bias
display(intercept, bias)
```

Agora aplicaremos em dois algoritmos iterativos. O primeiro com número de passos definidos e outro com a condição de parada quando a diferença entre cada iteração for insignificante

```python
intercept = 1
bias = 1.5

x = np.arange(1, 6)
y = [3, 1, 5, 4, 4]
learning_rate = 0.03

previsto = lambda intercept, bias: np.array([bias * i + intercept for i in x])
esperado = np.array(y)

residual_bias = lambda bias : np.square(previsto(intercept, bias) - esperado)
residual_intercept = lambda intercept : np.square(previsto(intercept, bias) - esperado)
```

```python
def plotar():
    
    plt.plot(x, previsto(intercept, bias))
    plt.scatter(x, y)
    plt.xlim(0, 7)
    plt.ylim(0,7)

    plt.title("Regressão Linear")
    plt.xlabel("Renda (em mil reais)")
    plt.ylabel("Quantidade de carros")

    plt.show()
```

```python
print("Intercept: ", intercept)
print("Bias: ", bias)
plotar()
for _ in range(11):
    print("Intercept: ", intercept)
    print("Bias: ", bias)
    gradiente_bias = np.gradient(residual_bias(bias))
    gradiente_intercept = np.gradient(residual_intercept(intercept))
    step_intercept = np.sum(gradiente_intercept) * learning_rate
    step_bias = np.sum(gradiente_bias) * learning_rate
    print()
    print("Passo intercep: ", step_intercept)
    print("Passo bias: ", step_bias)
    intercept -= step_intercept
    bias -= step_bias
    print()
    print("intercep: ", intercept)
    print("bias: ", bias)
    plotar()
```

```python
intercept = 1
bias = 1.5

x = np.arange(1, 6)
y = [3, 1, 5, 4, 4]
learning_rate = 0.03
limit = 1e-5


previsto = lambda intercept, bias: np.array([bias * i + intercept for i in x])
esperado = np.array(y)

residual_bias = lambda bias : np.square(previsto(intercept, bias) - esperado)
residual_intercept = lambda intercept : np.square(previsto(intercept, bias) - esperado)
```

```python
print("Intercept: ", intercept)
print("Bias: ", bias)
plotar()

loop = True
while loop:
    print("Intercept: ", intercept)
    print("Bias: ", bias)
    gradiente_bias = np.gradient(residual_bias(bias))
    gradiente_intercept = np.gradient(residual_intercept(intercept))
    step_intercept = np.sum(gradiente_intercept) * learning_rate
    step_bias = np.sum(gradiente_bias) * learning_rate
    print()
    print("Passo intercep: ", step_intercept)
    print("Passo bias: ", step_bias)
    intercept -= step_intercept
    bias -= step_bias
    print()
    print("intercep: ", intercept)
    print("bias: ", bias)
    plotar()
    if ((step_bias)**2 < (limit)**2 and (step_intercept)**2 < (limit)**2):
        loop = False
```

Não houve convergência de valores

Agora em vez de usar todos os valores, usaremos o gradiente descentente estotástico, que escolhe um valor por vez. Para isso, usaremos o SymPy

```python
import sympy as sp
```

Para regressão linear, temos:

$$previsto = ŷ = bias \cdot x + intercept$$
$$esperado = y$$

$$Erro = (ŷ - y)^2 = (bias \cdot x + intercept - y)^2$$
$$\frac{dErro}{dbias} = 2 \cdot x \cdot (bias \cdot x + intercept - y)$$
$$\frac{dErro}{dintercept} = 2 \cdot (bias \cdot x + intercept - y)$$

```python
intercept = 1
bias = 1.5

x = np.arange(1, 6)
y = [3, 1, 5, 4, 4]
learning_rate = 0.04

b, i, x0, y0 = sp.symbols('b i x0 y0')
loss = (b*x0 + i - y0)**2
residual_bias_symbol = sp.diff(loss,b)
residual_intercept_symbol = sp.diff(loss,i)
residual_bias = sp.lambdify((b, i, x0, y0), residual_bias_symbol, "numpy")
residual_intercept = sp.lambdify((b, i, x0, y0), residual_intercept_symbol, "numpy")
```

```python
print("Intercept: ", intercept)
print("Bias: ", bias)
plotar()
for _ in range(51):
    index = rd.randint(0, 4)
    x_ = x[index]
    y_ = y[index]
    print("Intercept: ", intercept)
    print("Bias: ", bias)
    step_intercept = residual_intercept(bias, intercept, x_, y_) * learning_rate
    step_bias = residual_bias(bias, intercept, x_, y_) * learning_rate
    print()
    print("Passo intercep: ", step_intercept)
    print("Passo bias: ", step_bias)
    intercept -= step_intercept
    bias -= step_bias
    print()
    print("intercep: ", intercept)
    print("bias: ", bias)
    plotar()
```

Essa técnica evita de encontrar todos os gradientes necessários, porém o gradiente descedente estotástico demora a convergir, por isso existe o chamdo mini-batch que escolhe uma quantidade menor de pontos e usando esses pontos ele usa o gradiente descendente

```python
intercept = 1
bias = 1.5

x = np.arange(1, 6)
y = np.array([3, 1, 5, 4, 4])
learning_rate = 0.04


previsto = lambda intercept, bias, x: np.array([bias * i + intercept for i in x])
esperado = lambda y : np.array(y)

residual_bias = lambda bias,x, y : np.square(previsto(intercept, bias, x) - esperado(y))
residual_intercept = lambda intercept,x, y : np.square(previsto(intercept, bias, x) - esperado(y))
```

```python
def plotar():
    
    plt.plot(x, previsto(intercept, bias, x))
    plt.scatter(x, y)
    plt.xlim(0, 7)
    plt.ylim(0,7)

    plt.title("Regressão Linear")
    plt.xlabel("Renda (em mil reais)")
    plt.ylabel("Quantidade de carros")

    plt.show()
```

```python
print("Intercept: ", intercept)
print("Bias: ", bias)
plotar()
for _ in range(16):
    pontos_escolhidos = np.random.choice(np.arange(0, 5), size=3, replace=False)
    new_x = np.array([x[i] for i in pontos_escolhidos])
    new_y = np.array([y[i] for i in pontos_escolhidos])
    print("Intercept: ", intercept)
    print("Bias: ", bias)
    gradiente_bias = np.gradient(residual_bias(bias, new_x, new_y))
    gradiente_intercept = np.gradient(residual_intercept(intercept, new_x, new_y))
    step_intercept = np.sum(gradiente_intercept) * learning_rate
    step_bias = np.sum(gradiente_bias) * learning_rate
    print()
    print("Passo intercep: ", step_intercept)
    print("Passo bias: ", step_bias)
    intercept -= step_intercept
    bias -= step_bias
    print()
    print("intercep: ", intercept)
    print("bias: ", bias)
    plotar()
```

O Mini-bach é mais rápido que o estocástico e consegue achar o mínimo usando menos computação que o gradiente descendente normal
