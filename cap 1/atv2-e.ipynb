Funções auxiliares

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

Função para plotar um gráfico

```python
def plotar():
    x = np.array([0, 0.5, 1])
    y = np.array([0, 1, 0])

    plt.scatter(x, y)
    plt.plot(np.arange(0, 1.01, 0.03), predict(np.arange(0, 1.01, 0.03)), color="green")
    plt.xlim(-0.03, 1.03)
    plt.ylim(-0.5,1.25)


    plt.title("Rede Neural")
    plt.xlabel("Dosagem da arma X")
    plt.ylabel("Eficácia")

    plt.show()
```

# Pesos fixos, variados e a taxa de aprendizado

```python
# Pesos fixos
w1 = np.float32(3.34)
w2 = np.float32(-1.53)

b1 = np.float32(-3.43)
b2 = np.float32(0.57)

# Pesos Variados
w3 = np.float64(0.5)   # peso entre h1 → y
w4 = np.float64(0.5)   # peso entre h2 → y
b3 = np.float64(0.0)   # bias da saída

# Resultado das funções sigmoid de cada neurônio
N1 = sigmoid(w1 * x + b1)
N2 = sigmoid(w2 * x + b2)

# Taxa de aprendizado
lr = 0.1
```

# Gradiente Descedente

<center>Considerando as funções de ativação</center>

$$N_1 = sigmoid(w_1 \cdot x + b_1) \\
N_2 = sigmoid(w_2 \cdot x + b_2)$$

<center>Sabendo que o previsto e dado pela função</center>

$$previsto = b_3 + w_3 \cdot N_1 + w_4 \cdot N_2
\\
erro = previsto - esperado \\
loss = erro^2$$

<center>Suas derivadas em função de b3, w3 e w4 podem ser expressas por:</center>

$$\frac{d_{loss}}{d_{b_3}} = \frac{d_{loss}}{d_{previsto}} \cdot \frac{d_{previsto}}{d_{b_3}} \\\
\\
\frac{d_{loss}}{d_{previsto}} = 2 \cdot erro
\\
\frac{d_{previsto}}{d_{b_3}} = 1 \\\
\\
\frac{d_{loss}}{d_{b_3}} = 2 \cdot erro \cdot 1\\\
\\
\frac{d_{loss}}{d_{w_3}} = \frac{d_{loss}}{d_{previsto}} \cdot \frac{d_{previsto}}{d_{w_3}} =  2 \cdot erro \cdot N_1 \\\
\\
\frac{d_{loss}}{d_{w_4}} = \frac{d_{loss}}{d_{previsto}} \cdot \frac{d_{previsto}}{d_{w_4}} =  2 \cdot erro \cdot N_2 $$

```python

# Dados iniciais
x = np.array([0, 0.5, 1])
y_esperado = np.array([0, 1, 0])

print("b3: ", b3)
print("w3: ", w3)
print("w4: ", w4)
plotar()

for epoch in range(10000):

    # y previsto pela rede
    y_previsto = b3 + w3 * N1 + w4 * N2

    # Função erro
    erro = y_previsto - y_esperado

    # gradientes
    db3 = (2) * np.sum(erro)
    dw3 = (2) * np.sum(erro * N1)
    dw4 = (2) * np.sum(erro * N2)

    # atualização dos valores com a taxa de aprendizado
    b3 -= lr * db3
    w3 -= lr * dw3
    w4 -= lr * dw4

    if epoch % 200 == 0:
        print("b3: ", b3)
        print("w3: ", w3)
        print("w4: ", w4)
        plotar()

print("b3: ", b3)
print("w3: ", w3)
print("w4: ", w4)
plotar()
```
